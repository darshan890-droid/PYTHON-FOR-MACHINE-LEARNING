```python
# Student scores
scores = [65, 72, 45, 80, 55, 60, 49, 90, 33, 77,
          58, 69, 74, 82, 47, 53, 61, 88, 92, 40]

# Average, highest, lowest
avg = sum(scores) / len(scores)
highest = max(scores)
lowest = min(scores)

print("Average Score:", avg)
print("Highest Score:", highest)
print("Lowest Score:", lowest)

# Pass / Fail
for s in scores:
    if s >= 50:
        print(s, "Pass")
    else:
        ("Fail")    

# Grade classification
for s in scores:
    if s >= 85:
        grade = "A"
    elif s >= 70:
        grade = "B"
    elif s >= 60:
        grade = "C"
    elif s >= 50:
        grade = "D"
    else:
        grade = "F"

    print(s, "Grade:", grade)
```
